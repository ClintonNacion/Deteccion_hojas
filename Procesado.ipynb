```python
import cv2
import os
import shutil
import numpy as np

# Rutas
input_folder = "/content/drive/MyDrive/Capstone/hoja_sana"
output_folder = "/content/drive/MyDrive/Capstone/hoja_sana_procesadas"

# Crear carpeta de salida limpia
if os.path.exists(output_folder):
    shutil.rmtree(output_folder)
os.makedirs(output_folder)

# Parámetros
blur_threshold = 40.0
target_size = (224, 224)

# Función para verificar borrosidad
def is_blurry(image, threshold=blur_threshold):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lap_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return lap_var < threshold

# Función para redimensionar con padding (mantiene aspecto)
def resize_with_padding(image, size=(224, 224)):
    h, w = image.shape[:2]
    scale = min(size[0]/h, size[1]/w)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(image, (new_w, new_h))

    delta_w = size[1] - new_w
    delta_h = size[0] - new_h
    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)

    # Padding con color negro
    color = [0, 0, 0]
    new_image = cv2.copyMakeBorder(resized, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
    return new_image

# Procesamiento
count = 0
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)
    image = cv2.imread(file_path)

    if image is None:
        continue

    if not is_blurry(image):
        padded = resize_with_padding(image, target_size)
        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, padded)
        count += 1

print(f"✅ Procesamiento completo Hojas Sanas. Imágenes guardadas: {count}")
```

##Repartir data propia y kaggle 80/20

```python
import os
import shutil
import random

random.seed(42)

# Configuración de rutas
base_propia = "/content/drive_mount/MyDrive/Capstone/dataset_propia"
base_kaggle = "/content/drive_mount/MyDrive/Capstone/dataset_kaggle"
dest_base = "/content/drive_mount/MyDrive/Capstone/dataset"

# Clases y su proporción
clases = ["sana", "minador", "phoma", "roya"]

for clase in clases:
    print(f"\n🔄 Procesando clase: {clase}")

    # Obtener archivos
    propia_imgs = [os.path.join(base_propia, clase, f) for f in os.listdir(os.path.join(base_propia, clase)) if f.endswith('.jpg')]
    kaggle_imgs = [os.path.join(base_kaggle, clase, f) for f in os.listdir(os.path.join(base_kaggle, clase)) if f.endswith('.jpg')]

    total_imgs = propia_imgs + kaggle_imgs
    random.shuffle(total_imgs)

    # Calcular cortes
    total_count = len(total_imgs)
    train_count = int(total_count * 0.8)

    train_imgs = total_imgs[:train_count]
    val_imgs = total_imgs[train_count:]

    # Función para mover
    def mover(imagenes, destino):
        for img in imagenes:
            filename = os.path.basename(img)
            dest_path = os.path.join(dest_base, destino, clase, filename)
            shutil.copy(img, dest_path)

    mover(train_imgs, "train")
    mover(val_imgs, "val")

    print(f"📁 Train: {len(train_imgs)} imágenes")
    print(f"📁 Val: {len(val_imgs)} imágenes")

print("\n✅ División completada con éxito.")
```

##Oversampling y data augmentación

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import numpy as np

base_train = "/content/drive_mount/MyDrive/Capstone/dataset/train"
clases = ["sana", "minador", "phoma", "roya"]

conteo = {
    "sana": 368,
    "minador": 37,
    "phoma": 65,
    "roya": 58
}

max_count = max(conteo.values())

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

for clase in clases:
    if conteo[clase] == max_count:
        continue  # No necesitamos aumentar esta clase

    carpeta = os.path.join(base_train, clase)
    imgs = [f for f in os.listdir(carpeta) if f.endswith('.jpg')]
    n_imgs = len(imgs)
    faltan = max_count - n_imgs
    print(f"\nClase {clase} tiene {n_imgs} imágenes. Generando {faltan} aumentadas...")

    # Cuántas nuevas imágenes generar por cada imagen original
    aug_por_img = faltan // n_imgs
    extra = faltan % n_imgs

    for i, img_file in enumerate(imgs):
        img_path = os.path.join(carpeta, img_file)
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)

        total_aug = aug_por_img + (1 if i < extra else 0)
        count = 0

        for batch in datagen.flow(
            x,
            batch_size=1,
            save_to_dir=carpeta,
            save_prefix=f"aug_{i}",
            save_format='jpg'
        ):
            count += 1
            if count >= total_aug:
                break

print("\n✅ Aumento de datos balanceado y diverso completado.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
# Configuración
train_dir = "/content/drive_mount/MyDrive/Capstone/dataset/train"
val_dir = "/content/drive_mount/MyDrive/Capstone/dataset/val"
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 15
NUM_CLASSES = 4
# Data generators con preprocesamiento MobileNetV2
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)
val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
# Modelo MobileNetV2 base
base_model = MobileNetV2(input_shape=(*IMG_SIZE, 3),
                         include_top=False,
                         weights='imagenet')
base_model.trainable = False  # Congelar base
# Modelo completo
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# Callbacks
checkpoint_cb = ModelCheckpoint(
    "mejor_modelo_mobilenetv2.h5",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max",
    verbose=1
)

earlystop_cb = EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    mode="max",
    verbose=1,
    restore_best_weights=True
)

reduce_lr_cb = ReduceLROnPlateau(
    monitor="val_accuracy",
    factor=0.5,
    patience=3,
    verbose=1,
    mode="max",
    min_lr=1e-7
)

# Entrenamiento inicial (con base congelada)
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=[checkpoint_cb, earlystop_cb, reduce_lr_cb]
)

# Fine-tuning: descongelar parte del base_model
base_model.trainable = True
fine_tune_at = 100
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

fine_tune_epochs = 10
total_epochs = EPOCHS + fine_tune_epochs

history_fine = model.fit(
    train_generator,
    epochs=total_epochs,
    initial_epoch=history.epoch[-1],
    validation_data=val_generator,
    callbacks=[checkpoint_cb, earlystop_cb, reduce_lr_cb]
)

```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50, EfficientNetB0, InceptionV3
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_resnet
from tensorflow.keras.applications.efficientnet import preprocess_input as preprocess_effnet
from tensorflow.keras.applications.inception_v3 import preprocess_input as preprocess_inception
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Rutas a tus carpetas de datos
train_dir = '/content/drive/MyDrive/Capstone/dataset/train'
val_dir = '/content/drive/MyDrive/Capstone/dataset/val'

batch_size = 32
num_classes = 4

# Generadores para ResNet50 y EfficientNet-B0 (224x224)
train_gen_224 = ImageDataGenerator(preprocessing_function=None)
val_gen_224 = ImageDataGenerator(preprocessing_function=None)

train_generator_224_resnet = train_gen_224.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator_224_resnet = val_gen_224.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # IMPORTANTE para matriz de confusión
)

# Generadores para InceptionV3 (299x299)
train_gen_299 = ImageDataGenerator(preprocessing_function=None)
val_gen_299 = ImageDataGenerator(preprocessing_function=None)

train_generator_299_inception = train_gen_299.flow_from_directory(
    train_dir,
    target_size=(299, 299),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator_299_inception = val_gen_299.flow_from_directory(
    val_dir,
    target_size=(299, 299),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # IMPORTANTE para matriz de confusión
)

def create_model(base_model_class, input_shape, preprocess_func):
    base_model = base_model_class(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )
    base_model.trainable = False

    inputs = layers.Input(shape=input_shape)
    x = preprocess_func(inputs) if preprocess_func else inputs
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    return model

def plot_confusion_matrix(model, val_generator, class_names, title):
    y_pred_prob = model.predict(val_generator)
    y_pred = np.argmax(y_pred_prob, axis=1)
    y_true = val_generator.classes
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(title)
    plt.show()

model_resnet = create_model(ResNet50, (224, 224, 3), preprocess_resnet)
model_effnet = create_model(EfficientNetB0, (224, 224, 3), preprocess_effnet)
model_inception = create_model(InceptionV3, (299, 299, 3), preprocess_inception)

for model in [model_resnet, model_effnet, model_inception]:
    model.compile(
        optimizer=optimizers.Adam(),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ReduceLROnPlateau(patience=3)
]

# Entrenar ResNet50 y mostrar matriz de confusión
print("Entrenando ResNet50...")
history_resnet = model_resnet.fit(
    train_generator_224_resnet,
    validation_data=val_generator_224_resnet,
    epochs=20,
    callbacks=callbacks
)
class_names_224 = list(val_generator_224_resnet.class_indices.keys())
plot_confusion_matrix(model_resnet, val_generator_224_resnet, class_names_224, "Matriz de Confusión ResNet50")

# Entrenar EfficientNet-B0 y mostrar matriz de confusión
print("Entrenando EfficientNet-B0...")
history_effnet = model_effnet.fit(
    train_generator_224_resnet,
    validation_data=val_generator_224_resnet,
    epochs=20,
    callbacks=callbacks
)
plot_confusion_matrix(model_effnet, val_generator_224_resnet, class_names_224, "Matriz de Confusión EfficientNet-B0")

# Entrenar InceptionV3 y mostrar matriz de confusión
print("Entrenando InceptionV3...")
history_inception = model_inception.fit(
    train_generator_299_inception,
    validation_data=val_generator_299_inception,
    epochs=20,
    callbacks=callbacks
)
class_names_299 = list(val_generator_299_inception.class_indices.keys())
plot_confusion_matrix(model_inception, val_generator_299_inception, class_names_299, "Matriz de Confusión InceptionV3")
```
